```python
# 4. Export/Move Model

import shutil
import os

# We might be in 'yolov5' directory now, so path to backend is ../../web_app/backend/
src = 'runs/train/waste_detection_model/weights/best.pt'
dst = '../../web_app/backend/best.pt'

if os.path.exists(src):
    # Ensure destination directory exists
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    shutil.copy(src, dst)
    print(f"✅ Model saved to {dst}")
else:
    print("⚠️ Training did not produce a model or path is incorrect.")
    print(f"Current CWD: {os.getcwd()}")
    if os.path.exists('runs/train'):
        print("Contents of runs/train:", os.listdir('runs/train'))
```
